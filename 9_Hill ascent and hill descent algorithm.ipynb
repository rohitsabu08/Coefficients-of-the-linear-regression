How will you modify this algorithm to find the maxima of the concave function(Hill Climbing/Ascent Algorithm)

```python
#libraries
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
# CREATING THE FUNCTION
def f(x):
    return(4*x**2-38*x+98)
```

```python
#DERIVATIVE OF THE FUNCTION

def df(x):
    return(8*x-38)
```

```python
#PLOTTING THE FUNCTION

p=np.arange(-10,20,0.25)
plt.plot(p,f(p),'-')

plt.show()
```

```python
#SETTING THE INITIAL VALUE OF X
x_ini=-10

#marking the initial value on the plot

p=np.arange(-10,20,0.25)
plt.plot(p,f(p),'-')      #plotting the function
plt.axvline(x_ini,color='r',linewidth=1)  #plotting the initial value
plt.show()
```

```python
import time
```

The hill descent algorithm to find the minimum

```python
#Try with the following step size : 0.01,0.05,0.11,0.20,0.24,0.26

x=-10    #initial value
count=0  #counter
alpha=0.2  #learning rate
tol = 1e-10
 
# Calculating the value of x which is approaching the minimum point
while np.abs(df(x))>tol: 
    count=count+1
    x=x-alpha*df(x)
    time.sleep(0.1)
    print((df(x),x))
    
print(x)
print(count) #gives the number of iterations
```

### Hill descent Algorithm

```python
# CREATING THE FUNCTION
def g(x):
    return(-4*x**2+38*x+98)
```

```python
# DERIAVTIVE OF THE FUNCTION
def dg(x):
    return(-8*x+38)
```

```python
# PLOTTING THE FUNCTION
p=np.arange(-10,20,0.25)
plt.plot(p,g(p),'-')

plt.show()
```

```python
# Calculating and plotting the value of x which is approaching the minimum point
x=-10    #initial value
count=0  #counter
alpha=0.01  #learning rate
tol = 1e-10
plt.plot(p,g(p),'-')

while np.abs(dg(x))>tol:
    count=count+1
    x=x+alpha*(dg(x))
    
    #time.sleep(0.5)
    plt.axvline(x,color='r',linewidth=1)
    #plt((df(x),x))
    
print(x)
print(count) #gives the number of iterations
```
